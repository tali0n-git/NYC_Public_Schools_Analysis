# Reassigning non-borough-named cities:

```python
# a function should be created to separate out non-borough-named cities, checking their addresses, assigning boroughs accordingly and dropping them if necessary
# eg. New York --> Manhattan, Jamaica --> Queens, etc.
```

```python
from thefuzz import process

def assign_borough(city_name):
    '''Assigns a NYC borough based on fuzzy matching of the input city name.
    If no good match is found, returns unchanged city_name with a printed error message.

    Input:
        city_name (str): The name of the city to match.
    Output:
        str: The matched borough name or the unchanged city_name, with a printed error message.
    '''

    # For future expansion, we can add more city names that map to boroughs

    boroughs = ['BROOKLYN', 'BRONX', 'MANHATTAN', 'QUEENS', 'STATEN ISLAND', 'NEW YORK', 'JAMAICA', 'BAYSIDE', 'JACKSON HEIGHTS']
    
    # Clean the input slightly for better accuracy
    input_str = city_name.upper().strip()
    
    # extractOne returns a tuple: (The Match, The Score)
    # Example: ("BROOKLYN", 95)
    match, score = process.extractOne(input_str, boroughs)
    
    # A score of 80 is usually the "sweet spot" for typos.
    # Below 60-70, it's likely a different word entirely.
    if score >= 73:
        if match == 'NEW YORK':
            return 'MANHATTAN'
        elif (match == 'JAMAICA') or (match == 'BAYSIDE') or (match == 'JACKSON HEIGHTS'):
            return 'QUEENS'
        else:
            return match
    
    print(f"Could not confidently match '{city_name}'. Best guess: '{match}' with score {score}%. Returning unchanged.")
    return city_name

# --- Test Cases ---
print(assign_borough("Brookln"))      # Output: BROOKLYN (Score ~93)
print(assign_borough("Manhatten"))    # Output: MANHATTAN (Score ~89)
print(assign_borough("Statin Is."))   # Output: STATEN ISLAND (Score ~73)
print(assign_borough("London"))       # Output: UNKNOWN (Score ~54)
```

```python
school_data['City'] = school_data['City'].apply(assign_borough)

school_data['City'].value_counts()
```
